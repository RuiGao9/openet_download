```python
from openet_download import download_et_timeseries

### ====== User Inputs === ###
YOUR_API_KEY = "APPLY FOR AN API KEY AT https://openetdata.org/"
model = "Ensemble"
longitude = -120.414164
latitude = 37.375004
interval = "daily"
dir_out = r"C:\GitHub\openet_download\outputs"
start_date = "2024-10-01"
end_date = "2025-09-30"
plot = True
show_plot = True

### ====== Keep Defaults Below === ###
out = download_et_timeseries(
    lon=longitude,
    lat=latitude,
    start=start_date,
    end=end_date,
    interval=interval,
    model = model,
    out_dir=dir_out,
    api_key=YOUR_API_KEY,
    plot=plot,
    show_plot=show_plot,
)
print(out["csv"], out["figure"])
print(out["data"].head())
```
